PART A: DATA LOADING AND BASIC UNDERSTANDING

```python
#1. Import the required Python libraries (pandas, numpy, matplotlib, seaborn).
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#2. Load the dataset into a pandas DataFrame.
df = pd.read_csv("Customer Purchasing Behaviors.csv")

#3.Display the number of rows and columns in the dataset.
print(df.shape)
print("Number of rows: ",df.shape[0])
print("Number of columns: ",df.shape[1])

#4.Print the column names of the dataset.
print(df.columns)

#5.Display the first 5 rows of the dataset.
df.head()
```

Pandas -> data handling and Data frames
numpy -> for numerical operations
matplotlib -> basic plotting
seaborn -> advanced statistical visualizations

df.shape returns a tuple consisting of (no of rows,no of columns)

PART B: DATA TYPES AND MISSING VALUES

```python
#6.Check the data type of each column.
print(df.info())

#7.Identify the number of missing values in each column.
print(df.isnull().sum())

#8.Fill missing values in numerical columns using the mean.


#9.Fill missing values in categorical columns using the mode

      
#10.Verify that there are no missing values remaining.
print(df.isnull().sum())
```

PART C: DESCRIPTIVE STATISTICS

```python
#11.Calculate the mean for all numerical columns.
print("\nThe mean values of the columns are: ")
print(df.mean(numeric_only=True))

# 12.Calculate the median for all numerical columns.
print("\nThe median values of the columns are: ")
print(df.median(numeric_only=True))

# 13.Calculate the standard deviation for all numerical columns.
print("\nThe Standard Deviation: ")
print(df.std(numeric_only=True))
      
# 14.Find the minimum and maximum values for numerical columns.
print(df.agg(['min','max']))

# 15.Generate a summary using describe()
print(df.describe())
```

#Mean : the average value of a datset.
-> represents the central value of data

#Median : middle value of a dataset.

#Mode : most frequently occurred value in a dataset.

#std : the stadard deviation measures how much the values in a dataset deviate from the mean

#agg(): The agg() function in pandas is used to apply one or more statistical operations to a DataFrame.


PART D: BASIC ANALYSIS AND VISUALISATION

```python
# #16.Create a histogram for the purchase amount column.
plt.figure()
plt.hist(df['purchase_amount'], bins=10)
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Amount')
plt.show()

# 17.Create a bar chart showing the count of customers by product category.
#since we dont have a product category we use the region attribute
region_counts = df['region'].value_counts()

plt.figure()
plt.bar(region_counts.index, region_counts.values)
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.title('Customer Count by Region')
plt.show()

# 18.Create a box plot for the purchase amount column.
plt.figure()
plt.boxplot(df['purchase_amount'])
plt.ylabel('Purchase Amount')
plt.title('Box Plot of Purchase Amount')
plt.show()

# 19.Create a scatter plot between age and purchase amount.
plt.figure()
plt.scatter(df['age'], df['purchase_amount'])
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.title('Age vs Purchase Amount')
plt.show()

# 20.Display a correlation heatmap for numerical columns.
corr = df.corr(numeric_only=True)

plt.figure()
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
