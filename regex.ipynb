1. In the regex created from r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group 0 cover? Group 1?
Group 2?

```python
import re
exp = re.search(r'(\d\d\d)-(\d\d\d-\d\d\d\d)','628-177-7859')
#Group-0
print(f"Group 0 covers:{exp.group(0)}")
#Group-1
print(f"Group 1 covers:{exp.group(1)}")
#Group-2
print(f"Group 2 covers:{exp.group(2)}")
```

3. Clean up dates in different date formats (such as 3/14/2015, 03-14-2015, and 2015/3/14) by
replacing them with dates in a single, standard format.

```python
import re

text = "Date formats are 3/14/2015, 03-14-2015, and 2015/3/14."

date_pattern = r'\d{1,4}[/-]\d{1,2}[/-]\d{1,4}'

dates = re.findall(date_pattern,text)

# print(dates)
def split_date(date):
  if '/' in date:
    parts = date.split('/')
  else:
    parts = date.split('-')
  return parts

def classify_date(parts):
  if len(parts[0]) == 4:
    return 'yyyy/mm/dd'
  elif len(parts[2]) == 4:
    return 'dd/mm/yyyy'

def normalize(parts):
  fmt = classify_date(parts)
  if fmt == 'yyyy/mm/dd':
    return f"{parts[2]}/{parts[1]}/{parts[0]}"
  else:
    return f"{parts[0]}/{parts[1]}/{parts[2]}"

transform = {}

for date in dates:
  parts = split_date(date)
  transform[date] = normalize(parts)

for old,new in transform.items():
  text = text.replace(old,new)

print(text)
```

6. How would you write a regex that matches a sentence where the first word is either Alice, Bob,
or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or
baseballs; and the sentence ends with a period? This regex should be case-insensitive.

```python
import re

text = '''Alice eats apples.
Bob pets cats.
Carol throws baseballs.
Alice throws Apples.
BOB EATS CATS.
RoboCop eats apples.
ALICE THROWS FOOTBALLS.
Carol eats7 cats.'''

match_pattern = re.findall(r'^(Alice|Bob|Carol) (eats|pets|throws) (apples|cats|baseballs)(\.)$',text,re.IGNORECASE | re.MULTILINE)
print(match_pattern)

```

7. Strong Password Detection Write a function that uses regular expressions to make sure the
password string it is passed is strong. A strong password is defined as one that is at least eight
characters long, contains both uppercase and lowercase characters, and has at least one digit.
You may need to test the string against multiple regex patterns to validate its strength.



```python
import re
passwords = "Ey0D{jI33Q , 12345678 , abcd1234 , Zeq71OK&i48"


def is_strong_password(password):
    if not re.search(r'.{8,}', password):
        return False
    if not re.search(r'[a-z]', password):
        return False
    if not re.search(r'[A-Z]', password):
        return False
    if not re.search(r'[0-9]', password):
        return False
    return True

for password in passwords.split():
  if is_strong_password(password):
    print(f"{password} is a strong password")
```
